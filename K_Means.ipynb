```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Örnek veri seti (2 boyutlu, 3 küme içeren rastgele veri)
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
```

```python
# K-Means modelini tanımla ve eğit
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Tahmin edilen küme etiketleri
labels = kmeans.labels_

# Küme merkezleri
centers = kmeans.cluster_centers_
```

```python
# Verileri küme renklerine göre çiz
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', label='Veri noktaları')

# Küme merkezlerini işaretle
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='x', label='Küme Merkezleri')

plt.title('K-Means Kümeleme')
plt.legend()
plt.show()
```

```python
# Yeni veri noktaları
new_data = np.array([[0, 0], [5, 5]])

# Kümeleri tahmin et
predicted_clusters = kmeans.predict(new_data)

print("Yeni verilerin tahmin edilen kümeleri:", predicted_clusters)
```

```python
inertia = []

# Farklı k değerleri için model eğitimi
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)  # Küme içi hata karesi toplamı

# Sonuçları görselleştir
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Yöntemi')
plt.xlabel('Küme Sayısı (k)')
plt.ylabel('Küme İçi Hata Karesi (Inertia)')
plt.show()
```
